```python
#Question 1
import pandas as pd
import numpy as np
import os.path as path
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
exercise_data_path = path.join(r"C:\Users\abbis\Downloads\Exercise_Data.csv")
exercise_data = pd.read_csv(exercise_data_path)
print(exercise_data.head())
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(exercise_data[['1 min', '15 min', '30 min']], cmap='rainbow', cbar=True)
plt.title('Heatmap of Pulse Data')
plt.xlabel('Time')
plt.ylabel('Entries')
plt.show()

```

```python
#After analyzing the heatmap, it can be concluded that the pulse rate increases as the time of exercise increases. 
```

```python
#pulse values by diet and by type of exercise
pulse_values = exercise_data.melt(id_vars=['id', 'diet', 'kind'], value_vars=['1 min', '15 min', '30 min'], var_name='Time', value_name='Pulse')
sns.catplot(x='diet', y='Pulse', hue='kind', data=pulse_values, kind='swarm', palette='Set2')
plt.title('Pulse Values by Diet and Exercise')
plt.xlabel('Diet')
plt.ylabel('Pulse (bpm)')
plt.show()

```

```python
#After analyzing the categorical plot, it can be concluded that the pulse rate is higher for the runners compared to 
# the walkers, and the pulse rate is higher for the high-fat diet compared to the low-fat diet.
```

```python
#Question 2
# Load the data
planets_data = sns.load_dataset('planets')
print(planets_data)
```

```python
# Relational plots

sns.scatterplot(x = 'method', y = 'distance', data = planets_data, color = 'purple')
plt.title('Method vs. Distance')
plt.xlabel('Method')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Distance')
plt.grid(axis='y')
plt.show()



sns.scatterplot(x='orbital_period', y='mass', data=planets_data, hue='method', style='method', palette='Set2')
plt.title('Orbital Period vs. Mass')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass')
plt.legend(title='Method')
plt.show()

```

```python
#The first scatter plot above shows the distance of planets discovered by different methods. It can be observed that the 
#'Radial Velocity' method has discovered planets at a greater distance compared to other methods.

#The second scatter plot above shows the relationship between the orbital period and mass of planets discovered by different methods.
# It can be observed that the 'Radial Velocity' method has discovered planets with higher mass and longer orbital periods.
```

```python
# Distributional plots
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(planets_data['distance'], bins=20, color='green', kde=True)
plt.title('Distance Distribution')
plt.xlabel('Distance')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)
sns.histplot(planets_data['mass'], bins=20, color='blue', kde=True)
plt.title('Mass Distribution')
plt.xlabel('Mass')
plt.ylabel('Frequency')
plt.show()
```

```python
#The first histogram above shows the distribution of the distance of planets discovered. It can be observed that most planets are
# discovered at a distance of around 0-1000 light years.

#The second histogram above shows the distribution of the mass of planets discovered. It can be observed that most planets have a mass
# between 0-10 Jupiter masses.
```

```python
# Categorical plots

sns.countplot(x='method', data=planets_data, palette='Set2')
plt.title('Method Count')
plt.xlabel('Method')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()


sns.boxplot(x='method', y='distance', data=planets_data, palette='Set2')
plt.title('Distance by Method')
plt.xlabel('Method')
plt.ylabel('Distance')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
#The count plot above shows the number of planets discovered by each method. It can be observed that the 'Radial Velocity' method
# has discovered the most number of planets.

#The box plot above shows the distribution of distance of planets discovered by each method. It can be observed that the 'Radial
# Velocity' method has discovered planets at a greater distance compared to other methods.
```
